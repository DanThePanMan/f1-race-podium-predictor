```python
import requests
from dotenv import load_dotenv
import os


url = "https://csfloat.com/api/v1/listings?limit=50&def_index=500,503,505,506,507,508,509,512,514,515,516,517,518,519,520,521,522,523,525,526&sort_by=most_recent"
load_dotenv()
headers = {
    "Authorization": os.getenv("CSFLOAT_API_KEY")
}
data = requests.get(url, headers=headers)

if data.status_code == 200:
    results = data.json()
else:
    print(f"Error: {data.status_code}")
    results = []



display(results['data'][0]['reference']['last_updated'])

# for item in results['data']:
#     display (item)



```

```python
from datetime import datetime, timezone, timedelta

all_listings = []
cursor = None
cutoff_time = datetime.now(timezone.utc) - timedelta(minutes=60)

while True:
    base_url = "https://csfloat.com/api/v1/listings?limit=50&def_index=500,503,505,506,507,508,509,512,514,515,516,517,518,519,520,521,522,523,525,526&sort_by=most_recent"
    if cursor:
        url = f"{base_url}&cursor={cursor}"
    else:
        url = base_url
    
    data = requests.get(url, headers=headers)
    
    if data.status_code != 200:
        print(f"Error: {data.status_code}")
        break
    results = data.json()
    
    
    for item in results['data']:
        last_updated_str = item['reference']['last_updated']
        last_updated = datetime.fromisoformat(last_updated_str.replace('Z', '+00:00'))
        if last_updated < cutoff_time:
            print(f"cutoff time reached{last_updated_str}")
            break
        all_listings.append(item)
    else:
        cursor = results.get('cursor')
        print(f"Fetched {len(results['data'])} listings. Total so far: {len(all_listings)}. Continuing...")
        continue
    
    break

print(f"\nTotal listings collected: {len(all_listings)}")
all_listings
    
```
